# Get reviews and join with non Fast Food Restaurants

```python
QUERY_rev = '''

SELECT 
r.user_id, r.business_id, r.stars as rev_stars, r.date, r.text,
r.useful, r.funny, r.cool,
b.business_id as bus_id, b.name, b.neighborhood, b.address, b.city, b.state, b.postal_code, 
b.latitude, b.longitude, b.stars as bus_stars, b.review_count, b.is_open, b.categories

FROM 
`nodal-reserve-227121.Yelp.yelp_reviews` r

LEFT JOIN
`nodal-reserve-227121.Yelp.yelp_business` b

ON
r.business_id = b.business_id

WHERE b.categories LIKE '%Rest%' 
AND b.categories NOT LIKE '%Fast%'
AND b.address NOT LIKE '%Airport%'

'''
query_job_rev = bigquery_client.query(QUERY_rev)
```

```python
df_reviews = query_job_rev.to_dataframe()
```
