## Random Forest Classifier

```python
cols = list(clf_df_encoded.drop(columns=['target','rev_stars']).columns)
y = clf_df_encoded.target.values
X = clf_df_encoded.drop(columns=['target','rev_stars']).values

X_train1, X_test1, y_train1, y_test1 = train_test_split(X, y, test_size=0.15, random_state=42)
clf_rand=RandomForestClassifier(n_estimators=20, 
                                max_depth=15,
                                class_weight={0:3,1:1}
                               )
clf_rand.fit(X_train1,y_train1)
#predicting
y_pred1 = clf_rand.predict(X_test1)
#Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm1 = confusion_matrix(y_test1, y_pred1)
y_prob1 = clf_rand.predict_proba(X_test1)[::,1]

#find important features
feature_imp = pd.Series(clf_rand.feature_importances_,index=cols).sort_values(ascending=False)
feature_imp[0:20]
```

```python
from sklearn.metrics import average_precision_score, precision_score, recall_score, precision_recall_curve

average_precision = average_precision_score(y_test1, y_pred1)
precision = precision_score(y_test1, y_pred1, average='weighted')
recall = recall_score(y_test1, y_pred1, average='weighted')

print('Average precision-recall score: {0:0.2f}'.format(
      average_precision))
print('Precision score: {0:0.2f}'.format(
      precision))
print('Recall score: {0:0.2f}'.format(
      recall))
```

```python
import matplotlib.pyplot as plt
from inspect import signature

precision, recall, _ = precision_recall_curve(y_test1, y_pred1)

step_kwargs = ({'step': 'post'}
               if 'step' in signature(plt.fill_between).parameters
               else {})
plt.step(recall, precision, color='b', alpha=0.2,
         where='post')
plt.fill_between(recall, precision, alpha=0.2, color='b', **step_kwargs)

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title('2-class Precision-Recall curve: AP={0:0.2f}'.format(
          average_precision))
```

```python
cm1
```

```python
fig, ax = plt.subplots(figsize = (8,8))

y_pred_proba_rand1 = y_prob1
fpr_rand1, tpr_rand1, _ = metrics.roc_curve(y_test1,  y_pred_proba_rand1)
auc_rand1 = metrics.roc_auc_score(y_test1, y_pred_proba_rand1)
plt.plot(fpr_rand1,tpr_rand1,label="data 1, auc="+str(auc_rand1))
plt.show()
```

## Surprise SVD Algorithm

```python
from surprise import Reader, Dataset
from surprise import SVD, accuracy
from surprise.model_selection import train_test_split
from surprise import accuracy
reader = Reader()
data = Dataset.load_from_df(clf_df_encoded[['user_id', 'business_id', 'rev_stars']], reader)
#split the data int train/test
trainset, testset = train_test_split(data, test_size=.15, random_state=42)
#use SVD to train the model
algo = SVD(n_factors=50, random_state=42)
algo.fit(trainset)
```

```python
svd_predictions = algo.test(testset)
accuracy.rmse(svd_predictions)
```

### Using GridSearchCV

```python
from surprise.model_selection import GridSearchCV

data = Dataset.load_from_df(clf_df_encoded[['user_id', 'business_id', 'rev_stars']], reader)
param_grid = {'n_epochs': [30, 40], 
              'lr_all': [0.004, 0.006],
              'reg_all': [0.18, 0.2, 0.24], 
              'n_factors' : [4, 8]}
gs = GridSearchCV(SVD, param_grid, measures=['rmse', 'mae'], cv=3)
gs.fit(data)
# best RMSE score
print(gs.best_score['rmse'])
# combination of parameters that gave the best RMSE score
print(gs.best_params['rmse'])
```

```python
algo_tuned = SVD(n_factors=8, 
                 lr_all=0.004,
                 reg_all=0.2,
                 n_epochs=30,
                 random_state=42)
algo_tuned.fit(trainset)
svd_tuned_predictions = algo_tuned.test(testset)
accuracy.rmse(svd_tuned_predictions)
```
